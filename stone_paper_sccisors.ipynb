```python
import random as rn
class Game(object):
    '''lets play stone paper sccisors'''
    def play(self)->str :
        '''stone paper sccisors GAME'''
        objects_list = ['st','sc','pa']
        round = int(input('how many rounds do you like to play :'))
        if round>=2 :
            cpu_score = 0
            human_score = 0 
            while True :
                if cpu_score == round or human_score == round :
                    break
                choose = input('whats your hand ?(st as stone sc as sccirors and pa as paper)')
                if choose in objects_list :
                    if choose ==  'st' :
                        cpu_choose = rn.choice(objects_list)
                        if cpu_choose == 'st' :
                            print(f'cpu choose = {cpu_choose}')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                            print('same hand !')
                        elif cpu_choose == 'pa' :
                            cpu_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('cpu won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                        else :
                            human_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('you won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                    if choose == 'sc' :
                        cpu_choose = rn.choice(objects_list)
                        if cpu_choose == 'sc' :
                            print(f'cpu choose = {cpu_choose}')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                            print('same hand !')
                        elif cpu_choose == 'pa' :
                            human_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('you won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                        else :
                            cpu_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('cpu won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                    if choose == 'pa' :
                        cpu_choose = rn.choice(objects_list)
                        if cpu_choose == 'pa' :
                            print(f'cpu choose = {cpu_choose}')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                            print('same hand !')
                        elif cpu_choose == 'st' :
                            human_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('you won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                        else :
                            cpu_score += 1
                            print(f'cpu choose = {cpu_choose}')
                            print('cpu won')
                            print(f'scores : you = {human_score} , cpu = {cpu_score}')
                else :
                    print('please enter a valid value (st as stone sc as sccirors and pa as paper)')
        else :
            print('at least 2 round needed ')
            return
        if cpu_score == round :
            print('cpu won the game')
            print(f'scores : you = {human_score} , cpu = {cpu_score}')
            return
        else :
            print('you won the game')
            print(f'scores : you = {human_score} , cpu = {cpu_score}')
            return
        
while ask:=input('wanna play :') :
        if ask[0]=='y' or ask[0]=='Y':
            game = Game()
            game.play()
        else :
            break
```
